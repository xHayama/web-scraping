```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd


url_google = ('https://www.google.com/')
dados_xbox_series_s= []
option = Options()
option.headless = True
print('abrindo o navegador...')
n = webdriver.Chrome(options = option)
print('acessando o site...')
n.get('https://www.google.com/search?q=xbox+series+s&tbm=shop&sxsrf=AOaemvJsAQxxu7Z0eo4M1Za1yrWsZAeZPw%3A1640812279567&ei=987MYeeRIoHX1sQP_sKugAg&oq=xbox+series+s&gs_lcp=Cgtwcm9kdWN0cy1jYxABGAAyBwgjELADECcyCAgAEIAEELADMggIABCABBCwAzIICAAQgAQQsAMyCAgAEIAEELADMggIABCABBCwAzIOCAAQgAQQsQMQgwEQsAMyCAgAEIAEELADMggIABCABBCwAzIOCAAQgAQQsQMQgwEQsAMyCwiuARDKAxCwAxAnMgsIrgEQygMQsAMQJzILCK4BEMoDELADECdKBAhBGAFQAFgAYNQVaANwAHgAgAEAiAEAkgEAmAEAyAENwAEB&sclient=products-cc')
sleep(3)
print('Coletando html...')
html_content= n.find_element(By.XPATH , '//*[@id="rso"]')
conteudo = html_content.get_attribute('outerHTML')
soup= BeautifulSoup(conteudo, 'html.parser')
lista_produtos = soup.findAll('div', class_="KZmu8e")
print ('Listando Produtos Encontrados...')
print('=-'*40)
try: 
    for dados in lista_produtos:

        produto = dados.find('div', class_="sh-np__product-title translate-content").text
        print (produto)
        preço= dados.find('b', class_="translate-content").text
        print(preço)
        loja= dados.find('span',  class_="E5ocAb").text
        print (loja)
        l= dados.find("a", class_="shntl sh-np__click-target")
        link= (url_google+l['href'])
        print(link)

        dados_xbox_series_s.append([produto, preço, loja, link])
        print('=-'*40)
except: 
    print('não foi possivel pegar o valor')

print('Salvando dados em um data frame...')
dados_xbox_series = pd.DataFrame(dados_xbox_series_s, columns =[ "produto", "preço", "loja", "link"])
dados_xbox_series.to_csv('xbox_series.csv', index=False)

print ('tarefa finalizada')
    
```
